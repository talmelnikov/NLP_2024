```python
import gensim
from gensim.models import Word2Vec

# Step 3: Store sentences in a list
corpus = [word_tokenize(sentence) for sentence in filtered_sentences]

# Print the corpus to verify
for idx, sentence in enumerate(corpus):
    print(f"Sentence {idx + 1}: {sentence}")
```

# Training Model

```python

# Define corpus (tokenize the sentences)

# Train the Word2Vec model
model = Word2Vec(
    sentences=corpus,      # The corpus to train the model on
    vector_size=100,       # The size of the word vectors to be learned
    window=5,              # The size of the window of words to be considered
    min_count=1,           # The minimum frequency required for a word to be included in the vocabulary
    sg=0,                  # 0 for CBOW, 1 for skip-gram
    negative=5,            # The number of negative samples to use for negative sampling
    ns_exponent=0.75,      # The exponent used to shape the negative sampling distribution
    alpha=0.03,            # The initial learning rate
    min_alpha=0.0007,      # The minimum learning rate to which the learning rate will be linearly reduced
    epochs=30,             # The number of epochs (iterations) over the corpus
    workers=4,             # The number of worker threads to use for training the model
    seed=42,               # The seed for the random number generator
    max_vocab_size=None    # The maximum vocabulary size (None means no limit)
)

# Save the model
model.save("word2vec_model")

# Load the model (for future use)
model = Word2Vec.load("word2vec_model")

# Example usage
word_vector = model.wv['war']
similar_words = model.wv.most_similar('war')
print("Vector for 'war':", word_vector)
print("Most similar words to 'war':", similar_words)
```

# Glove
GloVe (Global Vectors for Word Representation) creates word embeddings by analyzing word co-occurrence in a large text corpus. It produces dense vectors that capture word meanings and relationships. GloVe is effective for various NLP tasks and is available as an open-source tool.

```python
from gensim.models import KeyedVectors
from nltk.tokenize import word_tokenize
import gensim.downloader as api

# Load GloVe embeddings (using Gensim's downloader)
glove_vectors = api.load('glove-wiki-gigaword-100')  

# Process each sentence in the corpus
for idx, tokens in enumerate(corpus):
    # Get vectors for each token in the sentence
    vectors = [glove_vectors[word] for word in tokens if word in glove_vectors.key_to_index]
    
    # Print vectors for each token
    for token, vector in zip(tokens, vectors):
        print(f"Token: {token}, Vector shape: {vector.shape}, Vector: {vector}")
```

### Models like GloVe  represents:

Semantic Similarity: The vector encodes how words are used in context and their semantic relationships with other words.

Word Features: Each dimension in the vector represents a feature learned during the embedding process. These features are not directly interpretable but collectively capture various linguistic properties such as context, syntax, and semantics.

```python
import nltk
from nltk import word_tokenize

# Define a simple Context-Free Grammar (CFG)
grammar = nltk.CFG.fromstring("""
    S -> NP VP P
    NP -> PRP | DT NN | DT JJ NN
    VP -> VBD NP | VBP NP | VB NP | VB PP | VBG PP | VBP PP | VBN NP | VBN PP | VP PP | VBZ VP
    PP -> IN NP | IN JJ NP
    DT -> 'the' | 'a'
    PRP -> 'I'
    NN -> 'dog' | 'cat' | 'cookie'| 'channel' | 'thread'
    JJ -> 'big' | 'militant' | 'international' | 'former' | 'hostage'
    VBD -> 'saw' | 'ate' | 'walked'  'damaged' | 'defended'
    VBP -> 'are' | 'would' 
    VBN -> 'been' | 'ordered' | 'seen' | 'realized' 
    IN -> 'of' | 'about' | 'at' | 'on' | 'since' | 'through' | 'without' | 'following'
    Conj -> 'and' | 'but'
    P -> '.'
""")

# Define a function to perform POS tagging using CYK-like algorithm
def cyk_tag(sentence, grammar):
    words = word_tokenize(sentence)
    parser = nltk.ChartParser(grammar)
    try:
        for tree in parser.parse(words):
            tagged = tree.pos()
            return tagged
    except ValueError:
        return None

# Example sentences
sentences = [
    "I saw a dog.",
    "The cat ate a cookie in the park.",
    "I walked with the dog.",
    "He continues to work hard.",
    "The situation is intense.",
    "She ordered a coffee at the cafe."
]

# Perform tagging using CYK-like algorithm
for sentence in sentences:
    tagged = cyk_tag(sentence, grammar)
    if tagged:
        print(tagged)
    else:
        print(f"No parse tree found for: {sentence}")
```
