## nltk 

```python
def Nltk_remove_stopWord(words):
        stopwords = nltk.corpus.stopwords.words('english')
        filtered_tokens = [token.lower() for token in words if token.lower() not in stopwords and token.isalpha()]
        return filtered_tokens
    
def Nltk_Tokenitation(text):
    tokens = nltk.word_tokenize(text)
    return tokens
    
def Nltk_Lemmatization(words):
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    return lemmas
    
def nltk_stemming(words):
    stemmer = PorterStemmer()
    stems = [stemmer.stem(word) for word in words]
    return stems




```

```python
all_words1=' '.join(all_words)
nltk_token=Nltk_Tokenitation(all_words1)
print(nltk_token)
nltk_removed= Nltk_remove_stopWord(nltk_token)
print(nltk_removed)
nltk_lemma = Nltk_Lemmatization(nltk_removed)
print(nltk_lemma)
nltk_stem = nltk_stemming(nltk_lemma)
print(nltk_stem)
```

## spacy

```python
import spacy
from spacy.tokenizer import Tokenizer
from spacy.lang.en import English
from spacy.lang.en.stop_words import STOP_WORDS



def Spacy_Tokenitation(words):
    nlp = spacy.load("en_core_web_sm")    
    doc = nlp(words)
    tokens = [token.text.lower() for token in doc if not token.is_stop and token.is_alpha]
    return tokens


def Spacy_Lemmatization(words):
    nlp = spacy.load("en_core_web_sm")
    text = ' '.join(words)
    doc = nlp(text)
    lemmas = [token.lemma_ for token in doc]
    return lemmas

def Spacy_stemming(words):
    nlp = spacy.load("en_core_web_sm")
    stemmer = PorterStemmer()
    text = ' '.join(words)
    doc = nlp(text)
    stems = [stemmer.stem(token.text) for token in doc]
    return stems
 
```

```python

def Nltk_remove_stopWord_hebrew(words):
        stopwords = nltk.corpus.stopwords.words('hebrew')
        filtered_tokens = [token for token in words if token not in stopwords and is_hebrew(token)]
        return filtered_tokens
    
def Nltk_Tokenitation_hebrew(text):
    tokens = nltk.word_tokenize(text)
    return tokens
    
def Nltk_Lemmatization_hebrew(words):
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    return lemmas
    
def nltk_stemming_hebrew(words):
    stemmer = PorterStemmer()
    stems = [stemmer.stem(word) for word in words]
    return stems
    
def is_hebrew(word):
    # Unicode ranges for Hebrew script
    hebrew_ranges = [
        (0x0590, 0x05FF),  # Hebrew alphabet
        (0x05B0, 0x05C7),  # Hebrew vowel points
        (0x05D0, 0x05EA),  # Hebrew letters
        (0xFB1D, 0xFB4F),  # Hebrew ligatures
        (0x05F0, 0x05F4),]  # Hebrew punctuation
       # Iterate through each character in the word
    for char in word:
        char_code = ord(char)  # Get the Unicode code point of the character
        # Check if the character falls within any of the Hebrew ranges
        if any(start <= char_code <= end for start, end in hebrew_ranges):
            return True
    
    return False
```
